```python
import pandas as pd

transactions = pd.read_csv('Transactions.csv')
```

```python
customers = pd.read_csv('Customers.csv')
```

```python
products = pd.read_csv('Products.csv')
```

```python
transactions = pd.merge(transactions, products, on='ProductID')
```

```python
customer_transactions = pd.merge(transactions, customers, on='CustomerID')
```

```python
customers
```

```python
customer_transactions
```

```python
customer_transactions.drop(['Price_y'],axis=1,inplace=True)
```

```python
customer_transactions
```

```python
total_region=customer_transactions.groupby(['CustomerID','Region']).size().reset_index(name='Count')
```

```python
total_region = total_region.loc[total_region.groupby('CustomerID')['Count'].idxmax()]
```

```python
total_region = total_region[['CustomerID', 'Region']]
```

```python
total_region
```

```python
total_spending = customer_transactions.groupby('CustomerID')['TotalValue'].sum().reset_index()
total_spending.columns = ['CustomerID', 'TotalSpending']
```

```python
num_transactions = customer_transactions.groupby('CustomerID')['TransactionID'].count().reset_index()
num_transactions.columns = ['CustomerID', 'NumTransactions']
```

```python
favorite_category = customer_transactions.groupby(['CustomerID', 'Category']).size().reset_index(name='Count')
favorite_category = favorite_category.loc[favorite_category.groupby('CustomerID')['Count'].idxmax()]
favorite_category = favorite_category[['CustomerID', 'Category']]
favorite_category.columns = ['CustomerID', 'FavoriteCategory']
```

```python
customer_features = pd.merge(total_spending, num_transactions, on='CustomerID')
customer_features = pd.merge(customer_features, favorite_category, on='CustomerID')
customer_features=pd.merge(customer_features,total_region,on="CustomerID")
```

```python
customer_features
```

```python
customer_features = pd.get_dummies(customer_features, columns=['FavoriteCategory'], prefix='Category')
customer_features = pd.get_dummies(customer_features, columns=['Region'], prefix='region')
```

```python
customer_features
```

```python
from sklearn.metrics.pairwise import cosine_similarity


from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
customer_features_scaled = scaler.fit_transform(customer_features.iloc[:, 1:])


similarity_matrix = cosine_similarity(customer_features_scaled)
```

```python
import numpy as np


lookalike_recommendations = {}


customer_ids = customer_features['CustomerID'].values


for i in range(20):
    customer_id = customer_ids[i]
    similarity_scores = similarity_matrix[i]
    
    top_indices = np.argsort(similarity_scores)[-4:-1][::-1]
    top_customers = [(customer_ids[idx], similarity_scores[idx]) for idx in top_indices]
    
    lookalike_recommendations[customer_id] = top_customers
```

```python

lookalike_df = pd.DataFrame.from_dict(lookalike_recommendations, orient='index')
lookalike_df = lookalike_df.stack().apply(pd.Series).reset_index(level=1, drop=True)
lookalike_df.columns = ['LookalikeCustomerID', 'SimilarityScore']

lookalike_df.to_csv('Lookalike.csv', index_label='CustomerID')
```
