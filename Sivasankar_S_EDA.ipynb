#### The Transaction Table and Customer Table has the Customer Id let's merge together

```python
transaction_merge=transaction.merge(customer,on='CustomerID',how='left')
```

```python
transaction_merge.isnull().sum()
```

#### Transaction merge table has the product id and product table also has the product id Let's Merge it

```python
transaction_merge=transaction_merge.merge(products,on='ProductID',how='left')
```

```python
transaction_merge.isnull().sum()
```

#### Let'S Do it the Exploratory Data Analysis (EDA) 

#### Univariate Analysis

```python
plt.figure(figsize=(6, 5))
sns.histplot(transaction_merge['TotalValue'], kde=True, color='blue')
plt.title('Distribution of Total Transaction Value')
```

```python
transaction_merge.columns
```

```python
plt.figure(figsize=(10,2))
sns.boxplot(x=transaction_merge['Price_x'], color='green')
plt.title('Price Distribution')
```

```python
plt.figure(figsize=(8,4))
transaction_merge['Region'].value_counts().plot(kind='bar', color='purple')
plt.title('Region-wise Customer Distribution')
```

```python
plt.figure(figsize=(8,4))
transaction_merge['Category'].value_counts().plot(kind='bar', color='orange')
plt.title('Product Categories Distribution')
```

#### Bivariate Analysis

##### Total Value vs Price

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=transaction_merge['Price_x'], y=transaction_merge['TotalValue'], alpha=0.7, color='red')
plt.title('Price vs Total Transaction Value')
plt.xlabel('Price')
plt.ylabel('Total Value')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='TotalValue', data=transaction_merge, palette='viridis')
plt.title('Total Value by Region')
plt.xlabel('Region')
plt.ylabel('Total Value')
plt.xticks(rotation=45)
plt.show()
```

#### Correlation HeatMap

```python
corr = transaction_merge[['Price_x', 'Quantity', 'TotalValue']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

#### Multivariate Analysis

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='Region', y='TotalValue', hue='Category', data=transaction_merge, palette='mako')
plt.title('Total Value by Region and Category')
plt.ylabel('Total Sales (USD)')
plt.xticks(rotation=45)
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

#### Customer Loyality Analysis

```python
customer_transactions = transaction_merge.groupby('CustomerID')['TransactionID'].count()
repeat_customers = customer_transactions[customer_transactions > 1].count()
print(f"Number of repeat customers: {repeat_customers}")
```

#### OutLier Detection

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=transaction_merge['TotalValue'], color='red')
plt.title('Outliers in Total Transaction Value')
plt.show()
```

#### Regionwise Sales in the dataset


```python
region_wise=transaction_merge.groupby('Region')['TotalValue'].sum().sort_values(ascending=False)
```

```python
region_wise
```

# South America is the number one sales and Europe is the Number Two

#### Let's Which has the Top products by the sales

```python
top_p=transaction_merge.groupby('ProductName')['TotalValue'].sum().sort_values(ascending=False).head(5)
```

```python
top_p
```

#### Sales by the Category

```python
category_sales = transaction_merge.groupby('Category')['TotalValue'].sum().sort_values(ascending=False)
```

```python
category_sales
```

#### Monthly sales Trend in the Transaction Date given by the Month

```python
monthly_sales = transaction_merge.groupby(transaction_merge['TransactionDate'].dt.to_period('M'))['TotalValue'].sum()
```

```python
monthly_sales
```

#### Signup dates trends

```python
signup_rate=customer.groupby(customer['SignupDate'].dt.to_period('M')).size()
```

```python
signup_rate
```

#### Data Visualization 
#### Sales by the Region
